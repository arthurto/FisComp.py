```python
import matplotlib.pyplot as plt
import numpy as np
# import scienceplots
from scipy.interpolate import interp1d
```

```python
a,b = 0.0, np.pi

def f(x):
    return np.sin(x)**2

def f_int(x,a=a,b=b):
    x_ = [a,(a+b)/2,b]
    y = [f(a),f((a+b)/2),f(b)]
    return interp1d(x_,y,kind='quadratic')(x)
```

```python
x = np.linspace(a,b,100)
y = f(x)
y_int = f_int(x)
```

```python
plt.figure(figsize=(10,6))

plt.plot(x,y,label = "$f(x)$",color="#3b57f5")
plt.plot(x,y_int,label = "$f_{interp}(x)$",color="black")

plt.fill_between(x,y,color="#3b57f5",alpha=0.5)
plt.fill_between(x,y_int,color="black",alpha=0.5)

plt.xlim(a,b)
plt.ylim(0.0,1.2)

plt.xlabel("$x$",fontsize=20)
plt.ylabel("$f(x)$",fontsize=20)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)

plt.title("Interpolao de $f(x)$",fontsize=16)

plt.legend(fontsize=16)
plt.grid(alpha=0.4)
plt.show()
```
